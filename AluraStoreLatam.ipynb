### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tiendas_df = {
    "tienda": tienda,
    "tienda2": tienda2,
    "tienda3": tienda3,
    "tienda4": tienda4
}

ingresos_totales = {}
for nombre_tienda, df in tiendas_df.items():
    ingresos_totales[nombre_tienda] = df['Precio'].sum()

for nombre_tienda, ingresos in ingresos_totales.items():
    print(f"Ingresos totales de {nombre_tienda.replace('tienda', 'Tienda ')}: {ingresos:,.2f}")
```

# 2. Ventas por categoría

```python
ventas_por_categoria = {}
for nombre_tienda, df in tiendas_df.items():
    ventas_por_categoria[nombre_tienda] = df['Categoría del Producto'].value_counts()

for nombre_tienda, ventas_categoria in ventas_por_categoria.items():
    print(f"\nVentas por categoría en {nombre_tienda.replace('tienda', 'Tienda ')}:")
    print(ventas_categoria.head(5)) # Mostrar las 5 categorías más populares
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = {}
for nombre_tienda, df in tiendas_df.items():
    calificacion_promedio[nombre_tienda] = df['Calificación'].mean()

for nombre_tienda, calificacion in calificacion_promedio.items():
    print(f"\nCalificación promedio en {nombre_tienda.replace('tienda', 'Tienda ')}: {calificacion:.2f}")
```

# 4. Productos más y menos vendidos

```python
productos_vendidos_por_tienda = {}
for nombre_tienda, df in tiendas_df.items():
    productos_vendidos_por_tienda[nombre_tienda] = df['Producto'].value_counts()

for nombre_tienda, productos_counts in productos_vendidos_por_tienda.items():
    print(f"\n--- Productos más y menos vendidos en {nombre_tienda.replace('tienda', 'Tienda ')} ---")
    print("Productos más vendidos (Top 5):")
    print(productos_counts.head(5))
    print("\nProductos menos vendidos (Bottom 5):")
    print(productos_counts.tail(5))
```

# 5. Envío promedio por tienda

```python
costo_envio_promedio = {}
for nombre_tienda, df in tiendas_df.items():
    costo_envio_promedio[nombre_tienda] = df['Costo de envío'].mean()

for nombre_tienda, costo_envio in costo_envio_promedio.items():
    print(f"\nCosto de envío promedio en {nombre_tienda.replace('tienda', 'Tienda ')}: {costo_envio:,.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Gráfico 1: Ingresos Totales por Tienda
plt.figure(figsize=(10, 6))
sns.barplot(x=list(ingresos_totales.keys()), y=list(ingresos_totales.values()), hue=list(ingresos_totales.keys()), palette='viridis', legend=False)
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos Totales (COP)')
plt.xticks(ticks=range(len(ingresos_totales)), labels=[name.replace('tienda', 'Tienda ') for name in ingresos_totales.keys()])
plt.ticklabel_format(style='plain', axis='y') # Evita la notación científica en el eje Y
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Gráfico 2: Calificación Promedio por Tienda
plt.figure(figsize=(10, 6))
sns.barplot(x=list(calificacion_promedio.keys()), y=list(calificacion_promedio.values()), hue=list(calificacion_promedio.keys()), palette='coolwarm', legend=False)
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.xticks(ticks=range(len(calificacion_promedio)), labels=[name.replace('tienda', 'Tienda ') for name in calificacion_promedio.keys()])
plt.ylim(0, 5) # Las calificaciones suelen ir de 1 a 5
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Gráfico 3: Costo de Envío Promedio por Tienda
plt.figure(figsize=(10, 6))
sns.barplot(x=list(costo_envio_promedio.keys()), y=list(costo_envio_promedio.values()), hue=list(costo_envio_promedio.keys()), palette='plasma', legend=False)
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio (COP)')
plt.xticks(ticks=range(len(costo_envio_promedio)), labels=[name.replace('tienda', 'Tienda ') for name in costo_envio_promedio.keys()])
plt.ticklabel_format(style='plain', axis='y')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Para el Gráfico 4: Productos más vendidos por categoría (agregado de todas las tiendas)
# Primero combinamos los conteos de categorías de todas las tiendas
all_category_counts = pd.Series(dtype=int)
for store_name, counts in ventas_por_categoria.items():
    all_category_counts = all_category_counts.add(counts, fill_value=0)

# Obtenemos las 5 categorías más populares en general
top_5_overall_categories = all_category_counts.nlargest(5)

plt.figure(figsize=(12, 7))
sns.barplot(x=top_5_overall_categories.index, y=top_5_overall_categories.values, hue=top_5_overall_categories.index, palette='mako', legend=False)
plt.title('Top 5 Categorías de Productos Más Vendidas (Todas las Tiendas)')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45, ha='right') # Rota las etiquetas para mejor legibilidad
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Ajusta el diseño para evitar que las etiquetas se corten
plt.show()
```
